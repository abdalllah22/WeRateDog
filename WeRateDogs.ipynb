```python
t_archive_clean['name'] = t_archive_clean['name'].replace(regex=['^[a-z]+', 'None'], value=np.nan)
sum(t_archive_clean['name'].isnull())
```

```python
def parseNames(text):
    txt_list = text.split()
    for word in txt_list:
        if word.lower() == 'named':
            name_index = txt_list.index(word) + 1
            return txt_list[name_index]
        else:
            pass
    return np.nan
```

```python
t_archive_clean['name'] = np.where(t_archive_clean['name'].isnull(), t_archive_clean['text'].apply(parseNames), t_archive_clean['name'])
```

```python
sum(t_archive_clean['name'].isnull())
```

```python
stage_df = t_archive_clean['dog_stage'].value_counts()
stage_df
```

```python
#make a function fix_source which extract the strings between tags
def fix_source(x):
    'x is an html string from the source column in df_arch_cleaned dataset'
    #find the first closed  tag >
    i= x.find('>') + 1
    # find the first open tag after the previous <
    j =x[i:].find('<')
    # extract the text in between
    return x[i:][:j]
```

```python
t_archive_clean['source'] = t_archive_clean['source'].apply(lambda x: fix_source(x)).astype('category')
```

```python
plt.pie(
    stage_df,
    labels=['pupper','doggo','puppo','floofer'],
    autopct= '%1.1f%%',
    shadow=True,
    explode=(0.1,0.2,0.3,0.4) )
plt.title('Precentage of dog stage')
plt.axis('equal')
```

```python
data = t_archive_clean.groupby('source').count()['tweet_id']
ax = sns.barplot(y=data.index, x=data.values, palette='Blues_d')
ax.set(xlabel='Count', ylabel='Tweet source', title='Tweet Source Counts')
plt.show()
```
